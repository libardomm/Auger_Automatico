```python
def deduplicadorV2(diasDetectores):
    deduplicado=diasDetectores
    for i in range(len(diasDetectores)):
        for j in range(len(diasDetectores[0])):
            deduplicado[i][j]=diasDetectores[i][j][~diasDetectores[i][j].index.duplicated(keep='first')]        
    return deduplicado
```

```python
def concatenadorResampler(detCercanos):
    detec=[[] for i in range(len(detCercanos))]
    for i in range(len(detCercanos)):
        detec[i]=pd.concat(detCercanos[i]).resample("400s").ffill()
    
    concatenado=pd.concat(detec, axis=1)
    concatenado.drop("fLsId", axis=1, inplace=True) #Se borra todas las columnas con el nombre fLsId
    return concatenado

    
```

Se hace el remuestreo en aras de garantizar los 216 registros(1 registro cada 400 segundos) por dia y de esta manera, evitar que cada registro de cada detector corresponda con los registros de los otros.

La regla que se uso para hacer el remuestreo es el llenado hacia adelante, de modo que se no se usara informacion del futuro y asi seguir tratando los datos como una serie temporal.

# Leer Temperatura de la estación

Se cargan  los datos de una estación, y se remuestrean de modo que correspondan con la frecuencia con las que se remuestrearon los datos de los detectores. De la misma manera que con los datos de los detectores, estos tambien se remuestrearon usando la regla de llenado hacia delante.

```python
def lecturaTemperaturaStation(ruta, detectStation):
    nombres2=["Time","Temperature"]
    temperature_Station=pd.read_csv(str(ruta), sep=" ", header=None, names=nombres2, usecols=[0, 1], index_col=None)
    temperature_Station.set_index("Time", inplace=True)
    temperature_Station.index=pd.to_datetime(temperature_Station.index, unit="s")
    temperature_Station=temperature_Station.resample("400s").ffill()
    
    first=str(detectStation.index[0])
    last=str(detectStation.index[-1])
    temperature_Station=temperature_Station.loc[first:last]
    return temperature_Station
```

```python
def anulador_temperaturasBajas(temperature_Station):
    temperature_Station.loc[temperature_Station.Temperature<=3,"Temperature"]=None
    temperature_Station.ffill(inplace=True)
    return temperature_Station
```

```python
def mean_stDv( Trainfolds, data_stationA, data_stationB, option): #columnas train, filas test
    n=int(np.round(data_stationA.shape[0]/216))-Trainfolds
    estimador=selectStimator(option)
    regs=216
    errores=[]
    
    Xa,ya= XySplit(data_stationA)
    Xb, yb= XySplit(data_stationB)

    for n in range(0,n):
        f_day_n= regs*n #primer dia de entrenamiento
        l_day_n=f_day_n + (Trainfolds*regs) #ultimo dia de entrenan
        iX_train=Xa.iloc[f_day_n:l_day_n]
        iy_train=ya.iloc[f_day_n:l_day_n]

        iX_test= Xb.iloc[l_day_n: l_day_n+regs]
        iy_test=yb.iloc[l_day_n: l_day_n+regs]
        
        estimador.fit(iX_train,iy_train)
        iy_predic=estimador.predict(iX_test)

        errores.append((metrics.mean_absolute_error(iy_test,iy_predic)))
    return  str(round(np.mean(errores),2))+"(+/- "+str(round(np.std(errores),2))+")"
```

```python
def tablaPorConfiguracion(trainFold,estimador):
    data_Station_list=[data_LosLeones, data_LosMorados,data_LomaAmarilla,data_Coihueco]
    labels=["Los Leones", "Los Morados","Loma Amarilla","Coihueco"]
    my_df= []
    t=0

    for j in data_Station_list:

        d = {

            'Station': labels[t],
            'Los Leones' : mean_stDv(trainFold,data_Station_list[0],j,str(estimador)) ,
            'Los Morados' : mean_stDv(trainFold,data_Station_list[1],j,str(estimador)) ,
            'Loma Amarilla' : mean_stDv(trainFold,data_Station_list[2],j,str(estimador)) ,
            'Coihueco' : mean_stDv(trainFold,data_Station_list[3],j,str(estimador)) ,

        }
        t=t+1
        my_df.append(d)

    my_df = pd.DataFrame(my_df)

    my_df.set_index('Station', inplace=True)

    return my_df[["Los Leones","Los Morados", "Loma Amarilla","Coihueco"]]
```

```python
def mean_stDv4_Statations( Trainfolds, data_station1, data_station2,data_station3,data_station4, option):
    n=int(np.round(data_station1.shape[0]/216))-Trainfolds
    estimador=selectStimator(option)
    regs=216
    errores=[]
    
    X1,y1= XySplit(data_station1) # X train
    X2, y2= XySplit(data_station2)# y train
    X3,y3= XySplit(data_station3)#  X test
    X4, y4= XySplit(data_station4)# y test

    for n in range(0,n):
        f_day_n= regs*n #primer dia de entrenamiento
        l_day_n=f_day_n + (Trainfolds*regs) #ultimo dia de entrenan
        
        iX_train=X1.iloc[f_day_n:l_day_n]
        iy_train=y2.iloc[f_day_n:l_day_n]

        iX_test= X3.iloc[l_day_n: l_day_n+regs]
        iy_test=y4.iloc[l_day_n: l_day_n+regs]
        
        estimador.fit(iX_train,iy_train)
        iy_predic=estimador.predict(iX_test)

        errores.append((metrics.mean_absolute_error(iy_test,iy_predic)))
    return  str(round(np.mean(errores),2))+"(+/- "+str(round(np.std(errores),2))+")"
```

```python
def Combinaciones(trainFold,estimador):
    dsl=[data_LosLeones, data_LosMorados,data_LomaAmarilla,data_Coihueco]
    labels=["X=LL y=LL", "X=LL y=LM","X=LL y=LA","X=LL y=CH",
            "X=LM y=LL", "X=LM y=LM","X=LM y=LA","X=LM y=CH",
            "X=LA y=LL", "X=LA y=LM","X=LA y=LA","X=LA y=CH",
            "X=CH y=LL", "X=CH y=LM","X=CH y=LA","X=CH y=CH"]
    my_df= []
    t=0
    for i in dsl:
        for j in dsl:

            d = {
                'Combinaciones': labels[t],
                'X=LL y=LL' : mean_stDv4_Statations(trainFold,dsl[0],dsl[0],i,j,estimador) ,
                'X=LL y=LM' : mean_stDv4_Statations(trainFold,dsl[0],dsl[1],i,j,estimador) ,
                'X=LL y=LA' : mean_stDv4_Statations(trainFold,dsl[0],dsl[2],i,j,estimador) ,
                'X=LL y=CH' : mean_stDv4_Statations(trainFold,dsl[0],dsl[3],i,j,estimador) ,
                'X=LM y=LL' : mean_stDv4_Statations(trainFold,dsl[1],dsl[0],i,j,estimador) ,
                'X=LM y=LM' : mean_stDv4_Statations(trainFold,dsl[1],dsl[1],i,j,estimador) ,
                'X=LM y=LA' : mean_stDv4_Statations(trainFold,dsl[1],dsl[2],i,j,estimador) ,
                'X=LM y=CH' : mean_stDv4_Statations(trainFold,dsl[1],dsl[3],i,j,estimador) ,
                'X=LA y=LL' : mean_stDv4_Statations(trainFold,dsl[2],dsl[0],i,j,estimador) ,
                'X=LA y=LM' : mean_stDv4_Statations(trainFold,dsl[2],dsl[1],i,j,estimador) ,
                'X=LA y=LA' : mean_stDv4_Statations(trainFold,dsl[2],dsl[2],i,j,estimador) ,
                'X=LA y=CH' : mean_stDv4_Statations(trainFold,dsl[2],dsl[3],i,j,estimador) ,
                'X=CH y=LL' : mean_stDv4_Statations(trainFold,dsl[3],dsl[0],i,j,estimador) ,
                'X=CH y=LM' : mean_stDv4_Statations(trainFold,dsl[3],dsl[1],i,j,estimador) ,
                'X=CH y=LA' : mean_stDv4_Statations(trainFold,dsl[3],dsl[2],i,j,estimador) ,
                'X=CH y=CH' : mean_stDv4_Statations(trainFold,dsl[3],dsl[3],i,j,estimador) ,

            }
            t=t+1
            my_df.append(d)

    my_df = pd.DataFrame(my_df)

    my_df.set_index('Combinaciones', inplace=True)
    return my_df[labels]
```

```python
def mean_stDv_TEST( Trainfolds, data_stationA, data_stationB, option):
    n=int(np.round(data_stationA.shape[0]/216))-Trainfolds
    estimador=selectStimator(option)
    regs=216
    errores=[]
    
    y_test_list=[] 
    y_predict_list=[]  

    
    Xa,ya= XySplit(data_stationA)
    Xb, yb= XySplit(data_stationB)

    for n in range(0,n):
        f_day_n= regs*n #primer dia de entrenamiento
        l_day_n=f_day_n + (Trainfolds*regs) #ultimo dia de entrenan
        iX_train=Xa.iloc[f_day_n:l_day_n]
        iy_train=ya.iloc[f_day_n:l_day_n]

        iX_test= Xb.iloc[l_day_n: l_day_n+regs]
        iy_test=yb.iloc[l_day_n: l_day_n+regs]
        
        
       
        
        estimador.fit(iX_train,iy_train)
        iy_predic=estimador.predict(iX_test)
        
        y_test_list.append(iy_test)
        y_predict_list.append(iy_predic)

        errores.append((metrics.mean_absolute_error(iy_test,iy_predic)))
    return  y_test_list,y_predict_list, str(round(np.mean(errores),2))+"(+/- "+str(round(np.std(errores),2))+")"
```
